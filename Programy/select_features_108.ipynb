### Load libraries

```python
import random
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso

from collections import Counter
```

```python
# Load dataset
df_train_val = pd.read_csv('Csv/second.csv')
df_test = pd.read_csv('Csv/second_test.csv')
# Drop the first column by index
df_train_val.drop(df_train_val.columns[0], axis=1, inplace=True)
df_test.drop(df_test.columns[0], axis=1, inplace=True)
# Concatenate the DataFrames on top of one another
combined_df = pd.concat([df_train_val, df_test], ignore_index=True)
```

```python
lesion_label = 'category'

column_names = combined_df.columns.tolist()

feature_names = column_names[1:-1]

lesion_label = column_names[-1]
```

```python
lesion_X, lesion_y = combined_df[feature_names].values, combined_df[lesion_label].values

temp = list(zip(lesion_X, lesion_y))
random.shuffle(temp)
res1, res2 = zip(*temp)
# res1 and res2 come out as tuples, and so must be converted to lists.
lesion_X, lesion_y = list(res1), list(res2)  
```

```python
pipeline = Pipeline([
                     ('scaler',RobustScaler()),
                     ('model',Lasso())
])

search = GridSearchCV(pipeline,
                      {'model__alpha':np.arange(0.1,10,0.1)},
                      cv = 5, return_train_score=True, scoring="neg_mean_squared_error",verbose=3
                      )
```

```python
search.fit(lesion_X,lesion_y)
search.best_params_
coefficients = search.best_estimator_.named_steps['model'].coef_
importance = np.abs(coefficients)
np.array(feature_names)[importance > 0]
```

```python
selected = (np.array(feature_names)[importance > 0]).tolist()
selected.append('category')
```

```python
# Select only the columns in columns_to_keep
selected_train_val = df_train_val[selected]
selected_test = df_test[selected]
```

```python
# Save the filtered DataFrame to a CSV file
selected_train_val.to_csv('Csv/second_selected.csv', index=False)
selected_test.to_csv('Csv/second_test_selected.csv', index=False)
```
